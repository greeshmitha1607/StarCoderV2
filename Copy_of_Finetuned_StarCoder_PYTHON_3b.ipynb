```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer
from peft import PeftModel, PeftConfig

base_model = "bigcode/starcoder2-3b"
adapter_model = "Greeshmitha/finetunedPYTHON_starcoder2"

model = AutoModelForCausalLM.from_pretrained(base_model, torch_dtype=torch.bfloat16).to("cuda")
model = PeftModel.from_pretrained(model, adapter_model, torch_dtype=torch.bfloat16).to("cuda")
tokenizer = AutoTokenizer.from_pretrained(base_model)
```

```python
#Pass the input required for inference in this function.
# x = input string
# size = length of the desired output string

def inference(x, size):
  input = tokenizer.encode(x, return_tensors="pt").to('cuda')
  output = model.generate(input, max_length = size)
  text = tokenizer.decode(output[0])
  return text

print(inference(' Print reverse an array in python ', 100))
```
